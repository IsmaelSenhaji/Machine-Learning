## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = """select  DISTINCT LaunchSite from  SPACEXTBL;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
query = """select LaunchSite from  SPACEXTBL where LaunchSite like '%CCS%' limit 5;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = """select LaunchSite, Customers, sum(PayloadMass) from  SPACEXTBL where Customers like '%CRS%';"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """select LaunchSite, Customers, Serial, avg(PayloadMass) from  SPACEXTBL where Serial like '%B1%';"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """select LaunchSite, LandingOutcome, Date from  SPACEXTBL where LandingOutcome like '%True RTLS%'order by Date asc limit 1;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """select Serial, LandingOutcome, PayloadMass from  SPACEXTBL where LandingOutcome like '%True ASDS%' having PayloadMass between '4000' and '6000' order by PayloadMass desc ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = """select LaunchSite, Serial, PayloadMass, Count(Mission_Outcome)  from  SPACEXTBL group by Mission_Outcome  ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query = """select LaunchSite, Serial, PayloadMass  from  SPACEXTBL where PayloadMass > 1500  order by PayloadMass desc limit 10;"""
df  = pd.read_sql(query, con = engine )
df
  
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query = """select LaunchSite, Serial, Date, PayloadMass, LandingOutcome, Mission_Outcome  from  SPACEXTBL where LandingOutcome like '%None%' or LandingOutcome like '%False%' having Date = 2015;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 10

##### Muestra los resultados y la frecuencia de estos resultados entre 2010-06-04 y 2017-03-20 ordenados en orden descendiente según la frecuencia


```python
query = """select LaunchSite, Serial, Date, LandingOutcome, Mission_Outcome  from  SPACEXTBL  having Date between '2010-06-04' and '2017-03-20' order by Date desc ;"""
df  = pd.read_sql(query, con = engine )
df
```
